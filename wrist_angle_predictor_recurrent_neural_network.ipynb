```
import tensorflow as tf
import pandas as pd
import numpy as np
import math 
from tensorflow.keras import layers

emg=pd.read_csv('/content/new dataset.csv',)
emg=emg.drop(['Unnamed: 0'],1)
emgin=emg.drop(['output'],1)
```

```
emg_labels=tf.one_hot(emg['output'],depth=6)
sess=tf.Session()
with sess as se:
  emg_labels_one_hot=se.run(emg_labels)
```

```
col_max=[]
j=0
for i in emgin.columns:
  col_max.append(emgin[i].max())
  emgin[i]=emgin[i]/col_max[j]
  j=j+1
emgin
```

```
emg_seq=[]
seq_length=2
n=0

for i in range(0,(55080-seq_length+1)):
  emg_seq.append(np.array(emgin[n:n+seq_length]))
  n=n+1
emg_seq=np.array(emg_seq)

emg_labels=np.array(emg['output'])
emg_labels=emg_labels[(seq_length-1):55080]
```

```
model=tf.keras.Sequential([
layers.CuDNNLSTM(128,input_shape=(seq_length,7),return_sequences=True),
layers.CuDNNLSTM(128),
layers.Dense(6, activation='softmax')])
opt = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)
model.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(emg_seq[0:40000],emg_labels[0:40000],epochs=3)
results=model.evaluate(emg_seq[40000:45000],emg_labels[40000:45000])
print('test_loss,test_acc',results)
```
