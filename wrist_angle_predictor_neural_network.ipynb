```
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import pandas as pd
from sklearn import preprocessing
import math

n=0

emg = pd.read_csv('/content/25april 38400 delay afternoon features 2-edited.csv')
emg['O']=emg['O'].replace(30,1)
emg['O']=emg['O'].replace(60,2)
emg['O']=emg['O'].replace(90,3)
emg['O']=emg['O'].replace(120,4)
emg['O']=emg['O'].replace(150,5)
emgin=emg.drop(['O'],1)
#emgin=emgin[['A','H','I','J']]
emgin=np.array(emgin)
emg_labels=emg['O']




```

```
a=pd.DataFrame()
b=pd.DataFrame()
a=emg['A']
b=emg['O']
for i in ['B','C','D','E','F','G','H','I','J']:
   a=a.append(emg[i]).reset_index(drop=True)
   b=b.append(emg['O']).reset_index(drop=True)
c=pd.DataFrame({'reading':a ,'output':b})

```

```
d=[]
e=[]
f=[]
g=[]
h=[]
depth=6
sum_r=c.loc[0,'reading']
mean=c.loc[0,'reading']
rms=c.loc[0,'reading']
square_sum=(c.loc[0,'reading']*c.loc[0,'reading'])
initial_reading=c.loc[0,'reading']
dev_initial=0
d.append(mean)
e.append(rms)
f.append(sum_r)
g.append(square_sum)
h.append(dev_initial)
pervious_output=c.loc[0,'output']
count=1
for i in range(1,len(c)):
  #print(count)
  if c.loc[i,'output']==pervious_output:
    sum_r=sum_r+c.loc[i,'reading']
    square_sum=square_sum+c.loc[i,'reading']*c.loc[i,'reading']
    count=count+1
    mean=sum_r/count
    rms=math.sqrt(square_sum/count)
    dev_initial=c.loc[i,'reading']-initial_reading
    d.append(mean)
    e.append(rms)
    f.append(sum_r)
    g.append(square_sum)
    h.append(dev_initial)
    pervious_output=c.loc[i,'output']
  elif c.loc[i,'output']!=pervious_output:
    sum_r=c.loc[i,'reading']
    mean=c.loc[i,'reading']
    rms=c.loc[i,'reading']
    square_sum=c.loc[i,'reading']*c.loc[i,'reading']
    initial_reading=c.loc[i,'reading']
    dev_initial=0
    count=1
    d.append(mean)
    e.append(rms)
    f.append(sum_r)
    g.append(square_sum)
    h.append(dev_initial)
    pervious_output=c.loc[i,'output']

c['mean']=d
c['rms']=e
c['sum']=f
c['square_sum']=g
c['stddev']=c['mean']-c['reading']
c['dev_initial']=h
c.to_csv('new dataset.csv')

deapth=6
emg_labels=tf.one_hot(c['output'], depth)
sess=tf.Session()
with sess as se:
  emg_labels=se.run(emg_labels)

input1=c.drop(['output'],1)
```

```
col_max=[]
j=0
for i in input1.columns:
  col_max.append(input1[i].max())
  input1[i]=input1[i]/col_max[j]
  j=j+1
input1
```

```

input1_train=input1[0:40000]
emg_labels_train=emg_labels[0:40000]




model = tf.keras.Sequential([
layers.Dense(500, activation='relu',input_shape=(7,)), 
layers.Dense(500, activation='relu'),
layers.Dense(500, activation='relu'),
layers.Dense(6, activation='softmax')])

model.compile(optimizer=tf.train.AdamOptimizer(0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])



input1=np.array(input1)
model.fit(input1_train, emg_labels_train, epochs=8, batch_size=100)


```

```
input1_test=input1[40000:44000]
emg_labels_test=emg_labels[40000:44000]

results = model.evaluate(input1_test, emg_labels_test, batch_size=128)
print('test loss, test acc:', results)
```
